```python
df = df.drop_duplicates() #there are 101 duplicated rows which is removed by this function.
```

```python
df_salary_job_country = df.pivot_table(values = 'salary_year_avg', index = 'job_country', columns = 'job_title_short', aggfunc = 'median')
```

```python
top_countries = df['job_country'].value_counts().sort_values(ascending = False).head(10).index # top countries by most job posted
```

```python
top_countries
```

```python
top10_countries = df_salary_job_country.loc[top_countries]
```

```python
job_titles = ['Data Analyst', 'Data Engineer', 'Data Scientist']
```

```python
df_job_country_salary = top10_countries[job_titles]
```

```python
ax = df_job_country_salary.plot(kind = 'bar', figsize = (12,6))
ax.legend(loc = 'upper left')
plt.tight_layout()
plt.title('Job Posting Across Top 10 Countries')
plt.xlabel('Countries')
plt.ylabel('Total number of jobs')
plt.show()
#India and Germany has the highest median salary where as netherlands has highest median salary for data scientist
# The insights shows us that candidate needs to switch their jobs in respective countries to earn more income
```
